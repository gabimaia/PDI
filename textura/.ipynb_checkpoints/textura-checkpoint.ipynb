```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

im = cv2.imread('einstein.jpg',0)

Matrix_variancia = np.zeros(im.shape,np.uint8)
Matrix_media = np.zeros(im.shape, np.uint8)


w = 1              # raio do pixel vizinho
ny = len(im)
nx = len(im[0])


for i in range(w,nx-w):
    for j in range(w,ny-w):

        amostra = im[j-w:j+w, i-w:i+w].flatten()

        Matrix_variancia[j][i] = np.var(amostra)
        Matrix_media[j][i] = 255*(np.mean(amostra)>100)
        #matriz_skew[j][i] = skew(amostra)
        #matriz_kurtosis = kurtosis(amostra)
        
```

```python
plt.figure(figsize=(10,10))

plt.axis('off')

plt.subplot(121),plt.imshow(Matrix_variancia, cmap = 'gray')
plt.title('filtro variância')
plt.subplot(122),plt.imshow(Matrix_media, cmap = 'gray')
plt.title('filtro media')
plt.show()
```
